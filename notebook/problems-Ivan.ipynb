## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Normal Distribution


```python


from scipy.stats import norm

def dnorm(x, mean=0, sd=1):
    """
    Calculates the point estimate of the normal distribution (PDF).
    """
    result = norm.pdf(x, loc=mean, scale=sd)
    return result

def pnorm(q, mean=0, sd=1):
    """
    Calculates the cumulative probability of the normal distribution (CDF).
    """
    result = norm.cdf(q, loc=mean, scale=sd)
    return result

def qnorm(p, mean=0, sd=1):
    """
    Calculates the quantile function from the normal distribution (PPF).
    """
    result = norm.ppf(p, loc=mean, scale=sd)
    return result

def rnorm(n, mean=0, sd=1):
    """
    Generates random variables from the normal distribution.
    """
    result = norm.rvs(loc=mean, scale=sd, size=n)
    return result
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Parameters for Normal Distribution
mean = 0
sd = 1

# Generate x values and probabilities
x = np.linspace(-4, 4, 500)  # For PDF and CDF
quantiles = np.linspace(0.01, 0.99, 100)  # For PPF

# Compute PDF, CDF, and PPF using custom functions
pdf = dnorm(x, mean=mean, sd=sd)
cdf = pnorm(x, mean=mean, sd=sd)
ppf = qnorm(quantiles, mean=mean, sd=sd)


# Plotting
plt.figure(figsize=(12, 8))

# PDF
plt.subplot(3, 1, 1)
plt.plot(x, pdf, label="PDF (dnorm)", color='blue')
plt.title("Normal Distribution - PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid()
plt.legend()

# CDF
plt.subplot(3, 1, 2)
plt.plot(x, cdf, label="CDF (pnorm)", color='green')
plt.title("Normal Distribution - CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.legend()

# PPF
plt.subplot(3, 1, 3)
plt.plot(quantiles, ppf, label="PPF (qnorm)", color='red')
plt.title("Normal Distribution - PPF")
plt.xlabel("Probability")
plt.ylabel("Quantile")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

### Uniform Distribution

```python
from scipy.stats import uniform

def dunif(x, min=0, max=1):
    """
    Calculates the point estimate of the uniform distribution (PDF).
    """
    result = uniform.pdf(x, loc=min, scale=max-min)
    return result

def punif(q, min=0, max=1):
    """
    Calculates the cumulative probability of the uniform distribution (CDF).
    """
    result = uniform.cdf(q, loc=min, scale=max-min)
    return result

def qunif(p, min=0, max=1):
    """
    Calculates the quantile function from the uniform distribution (PPF).
    """
    result = uniform.ppf(p, loc=min, scale=max-min)
    return result

def runif(n, min=0, max=1):
    """
    Generates random variables from the uniform distribution.
    """
    result = uniform.rvs(loc=min, scale=max-min, size=n)
    return result
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Parameters for Uniform Distribution
min_val = 0
max_val = 1

# Generate x values and probabilities
x = np.linspace(-4, 4, 100)  # For PDF and CDF
quantiles = np.linspace(0.01, 0.99, 100)  # For PPF

# Compute PDF, CDF, and PPF using custom functions
pdf = dunif(x, min=min_val, max=max_val)
cdf = punif(x, min=min_val, max=max_val)
ppf = qunif(quantiles, min=min_val, max=max_val)

# Plotting
plt.figure(figsize=(12, 8))

# PDF
plt.subplot(3, 1, 1)
plt.plot(x, pdf, label="PDF (dunif)", color='blue')
plt.title("Uniform Distribution - PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid()
plt.legend()

# CDF
plt.subplot(3, 1, 2)
plt.plot(x, cdf, label="CDF (punif)", color='green')
plt.title("Uniform Distribution - CDF")
plt.xlabel("x")
plt.ylabel("Cumulative Probability")
plt.grid()
plt.legend()

# PPF
plt.subplot(3, 1, 3)
plt.plot(quantiles, ppf, label="PPF (qunif)", color='red')
plt.title("Uniform Distribution - PPF")
plt.xlabel("Probability")
plt.ylabel("Quantile")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
